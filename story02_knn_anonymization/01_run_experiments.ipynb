# Story 02 — Run All Iterations

Executes the full experiment grid. Each iteration:
1. Applies k-anonymity (k=5)
2. Replaces suppressed rows via mixed-variable KNN
3. Reports TVD, KS, utility, privacy, runtime
4. Saves outputs to `iterations/<meaningful_folder_name>/`


```python
import sys
from pathlib import Path
import pandas as pd

STORY_ROOT = Path(".").resolve()
sys.path.insert(0, str(STORY_ROOT))

from knn_lib import load_dataset, run_all_experiments, EXPERIMENT_GRID, ITERATIONS_DIR

df = load_dataset()
print(f"Dataset: {len(df):,} rows")
print(f"Iterations to run: {len(EXPERIMENT_GRID)}")
print(f"Output root: {ITERATIONS_DIR}")
```

```python
ranking = run_all_experiments(df)
ranking
```

```python
print("\nIteration folders created:")
for p in sorted(ITERATIONS_DIR.iterdir()):
    if p.is_dir():
        files = list(p.glob("*"))
        print(f"  {p.name}/  ({len(files)} files)")
```

## Outputs per iteration folder

| File | Content |
|------|---------|
| `config.json` | Run configuration |
| `anonymized_dataset.csv` | Dataset after k-anon replacement |
| `category_metrics.csv` | TVD per categorical column |
| `numeric_metrics.csv` | KS per numerical column |
| `scorecard.csv` | Pass/fail per metric area |
| `summary.json` | Runtime, pass rates, F1, suppression stats |

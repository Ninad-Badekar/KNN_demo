# Exploratory Data Analysis — Bank Customer Churn (No separate targets)

**Dataset:** `Bank Customer Churn Prediction.csv`  
All columns including `churn` are treated as regular categorical/numerical features.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

FILE_PATH = "Bank Customer Churn Prediction.csv"
ID_COLS = ["customer_id"]
CATEGORICAL_COLS = ["country", "gender", "churn"]
NUMERICAL_COLS = [
    "credit_score", "age", "tenure", "balance",
    "products_number", "credit_card", "active_member", "estimated_salary",
]
UTILITY_COL = "churn"  # optional label for relationship / utility-style plots
```

## 1. Load & first look


```python
df = pd.read_csv(FILE_PATH)
print("Shape:", df.shape)
print("\nCategorical:", CATEGORICAL_COLS)
print("Numerical:", NUMERICAL_COLS)
df.head()
```

## 2. Data quality


```python
info = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str).values,
    "missing": df.isna().sum().values,
    "missing_pct": (100 * df.isna().mean()).round(2).values,
    "n_unique": df.nunique().values,
})
print(info.to_string(index=False))
print(f"\nDuplicate rows: {df.duplicated().sum()}")
```

## 3. Categorical distributions (incl. churn)


```python
fig, axes = plt.subplots(1, len(CATEGORICAL_COLS), figsize=(5 * len(CATEGORICAL_COLS), 4))
if len(CATEGORICAL_COLS) == 1:
    axes = [axes]
for ax, col in zip(axes, CATEGORICAL_COLS):
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, ax=ax, palette="pastel")
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 4. Numerical distributions


```python
n = len(NUMERICAL_COLS)
ncols = 4
nrows = int(np.ceil(n / ncols))
fig, axes = plt.subplots(nrows, ncols, figsize=(4 * ncols, 3 * nrows))
for ax, col in zip(np.atleast_1d(axes).ravel(), NUMERICAL_COLS):
    sns.histplot(df[col], kde=True, bins=30, ax=ax, color="steelblue")
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 5. Correlation heatmap


```python
plot_cols = NUMERICAL_COLS + [UTILITY_COL] if UTILITY_COL in CATEGORICAL_COLS else NUMERICAL_COLS
corr = df[plot_cols].apply(pd.to_numeric, errors="ignore").select_dtypes(include="number").corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlations (numerical + utility column)")
plt.tight_layout()
plt.show()
```
